```python
# Funcion del menu
def menu():
        print('Sistema de Control de Cursos\nSeleccione acción a realizar\n1. Agregar curso\n2. Agregar calificación a curso\n3. Mostrar promedio por curso\n4. Mostrar curso con mejor promedio\n5. Listar todos los cursos\n6. Salir')
while True:
     menu()   
# Funcionalidad de las opciones del menu
     opcion = int(input('Ingrese opción: '))
     match opcion:
            case 1:
                agregarCurso()
            case 2:
                agregarCalificacion()
            case 3:
                mostrarPromedio()
            case 4:
                mostrarMejorPromedio()
            case 5:
                listarCursos()
            case 6:
                print('Saliendo del programa...')
     break



```

```python
# Declarar diccionario y rellenar por input sin funciones
def agregarCurso():
    cursos={}
    nombreCurso = input('Ingrese el nombre del curso: ')
    cursos[nombreCurso] = []
    print(f'Curso {nombreCurso} agregado exitosamente')
    return cursos

agregarCurso()
```
